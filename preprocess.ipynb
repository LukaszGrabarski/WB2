# Translation

## Translating train

```python
translator = Translator()

# Initialize the counter
counter = 0

def translate_text(text):
    global counter
    try:
        translated = translator.translate(text, src='auto', dest='en')
        counter += 1
        if counter % 100 == 0:  # print the counter every 100 translations
            print(f'Translated {counter} texts')
        return translated.text
    except Exception as e:
        return text  # In case of an error, return the original text

# Apply the function to the 'premise' and 'hypothesis' columns
train['premise'] = train['premise'].apply(translate_text)
train['hypothesis'] = train['hypothesis'].apply(translate_text)
```

# Removing stop words, tokenization for translated

```python
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split

# Ensure required resources are downloaded
nltk.download('stopwords')
nltk.download('punkt')

# Load the data
train_df = pd.read_csv('data/train_translated.csv')
test_df = pd.read_csv('data/test_translated.csv')

# Define stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Tokenize the text
    words = nltk.word_tokenize(text)
    
    # Remove stopwords and punctuation
    words = [word for word in words if word not in stop_words and word not in punctuation]
    
    # Join the words back into a single string
    text = ' '.join(words)
    
    return text

# Apply the preprocessing function to the premise and hypothesis columns
train_df['premise'] = train_df['premise'].apply(preprocess_text)
train_df['hypothesis'] = train_df['hypothesis'].apply(preprocess_text)

test_df['premise'] = test_df['premise'].apply(preprocess_text)
test_df['hypothesis'] = test_df['hypothesis'].apply(preprocess_text)
```

# Removing stop words, tokenization for not translated

```python
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split

# Ensure required resources are downloaded
nltk.download('stopwords')
nltk.download('punkt')

# Load the data
train_df = pd.read_csv('data/train.csv')
test_df = pd.read_csv('data/test.csv')

# Define stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Tokenize the text
    words = nltk.word_tokenize(text)
    
    # Remove stopwords and punctuation
    words = [word for word in words if word not in stop_words and word not in punctuation]
    
    # Join the words back into a single string
    text = ' '.join(words)
    
    return text

# Apply the preprocessing function to the premise and hypothesis columns
train_df['premise'] = train_df['premise'].apply(preprocess_text)
train_df['hypothesis'] = train_df['hypothesis'].apply(preprocess_text)

test_df['premise'] = test_df['premise'].apply(preprocess_text)
test_df['hypothesis'] = test_df['hypothesis'].apply(preprocess_text)
```

```python
train_df_not_translated = train_df.copy()
```

# Training translated

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

train_df = train_df_translated.copy()
test_df = test_df_translated.copy()

# Create a test_new dataframe by sampling from the train_df
test_new_df = train_df.sample(frac=0.2, random_state=42).reset_index(drop=True)
train_new_df = train_df.drop(test_new_df.index).reset_index(drop=True)

# Now, let's split the train_new_df into training and validation sets
train_texts, val_texts, train_labels, val_labels = train_test_split(
    train_new_df[['premise', 'hypothesis']], train_new_df['label'], test_size=0.2, random_state=42
)

# Convert text pairs into single strings for vectorization
train_texts_combined = train_texts['premise'] + ' ' + train_texts['hypothesis']
val_texts_combined = val_texts['premise'] + ' ' + val_texts['hypothesis']
test_texts_combined = test_new_df['premise'] + ' ' + test_new_df['hypothesis']

# Logistic Regression Model
vectorizer = TfidfVectorizer(max_features=5000)
X_train = vectorizer.fit_transform(train_texts_combined)
X_val = vectorizer.transform(val_texts_combined)
X_test = vectorizer.transform(test_texts_combined)

log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, train_labels)
val_preds_log_reg = log_reg.predict(X_val)
test_preds_log_reg = log_reg.predict(X_test)

# Evaluate Logistic Regression Model
val_accuracy_log_reg = accuracy_score(val_labels, val_preds_log_reg)
val_report_log_reg = classification_report(val_labels, val_preds_log_reg)
test_accuracy_log_reg = accuracy_score(test_new_df['label'], test_preds_log_reg)
test_report_log_reg = classification_report(test_new_df['label'], test_preds_log_reg)

print(f"Logistic Regression Validation Accuracy: {val_accuracy_log_reg}")
print(f"Logistic Regression Validation Report:\n{val_report_log_reg}")
print(f"Logistic Regression Test Accuracy: {test_accuracy_log_reg}")
print(f"Logistic Regression Test Report:\n{test_report_log_reg}")

results = {
    "Logistic Regression": {
        "Validation Accuracy": val_accuracy_log_reg,
        "Test Accuracy": test_accuracy_log_reg,
        "Test Report": test_report_log_reg
    }
}

results
```
